```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train,y_train)
```

```python
y_pred = mlr.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

# Backward Elimination Method

```python
import statsmodels.api as sm
```

```python
x
```

```python
x = np.append(arr= np.ones((50,1)).astype(int),values=x,axis=1) #axis=1 is for column
# To the 1's column append the x columns so constant(a0) col will be at 1st position.
```

```python
x
```

```python
x.shape
```

```python
x_opt = np.array(x[:,[0,1,2,3,4,5]],dtype=float)
regression_ols = sm.OLS(endog=y,exog=x_opt).fit()
regression_ols.summary()
```

```python
x_opt = np.array(x[:,[0,1,3,4,5]],dtype=float)
regression_ols = sm.OLS(endog=y,exog=x_opt).fit()
regression_ols.summary()
```

```python
#Now x1 have high p value. Remove x1 column.
x_opt = np.array(x[:,[0,3,4,5]],dtype=float)
regression_ols = sm.OLS(endog=y,exog=x_opt).fit()
regression_ols.summary()
```

```python
#Now x2 have high p value which is actually x4,  Remove x4 column.
x_opt = np.array(x[:,[0,3,5]],dtype=float)
regression_ols = sm.OLS(endog=y,exog=x_opt).fit()
regression_ols.summary()
```
