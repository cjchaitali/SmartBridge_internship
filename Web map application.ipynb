### Adding points from file

```python
import folium
import pandas as pd
df= pd.read_csv("D:/Udemy/Python/app2_web_map/Volcanoes.csv")
df
lat= list(df["LAT"])
lon= list(df["LON"])
map= folium.Map(Location=[80,-100],zoom_start=6)
#map.save("Map1.html")
fg = folium.FeatureGroup(name="My Map")

for lt,ln in zip(lat,lon):
    fg.add_child(folium.Marker(location=[lt,ln], popup="Hi i am marker",icon=folium.Icon(color="green")))
map.add_child(fg)
map.save("Map1.html")
```
